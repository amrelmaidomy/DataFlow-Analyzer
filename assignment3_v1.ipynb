# Write a Python program to removes multiple items from a given set

## Example
my_set: {1, 2, 3, 4, 5}

items to be removed: {1, 4}

Output: {2, 3, 5}

```python
# Write your answer here


my_set = {1, 2, 3, 4, 5}
items_to_remove = {1, 4}

my_set.difference_update(items_to_remove)

print(my_set)  
```

# find the factorial of a number 
## Example
(factorial of 5) = 5 * 4 * 3 * 2 * 1 = 120

```python
# Write your answer here

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

# Example
num = int(input("Enter a number: "))
print(f"factorial of {num} = {factorial(num)}")
```

# Create a List of Squares of Numbers from 1 to 10
## Example
Output: [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]

```python
# Write your answer here

# Create an empty list
squares = []

# Generate squares from 1 to 10
for i in range(1, 11):
    squares.append(i ** 2)

# Display the result
print(squares)
```

# Create a function that Calculates the Sum of Even Numbers in a List
## Example
numbers: [1, 2, 3, 4, 5, 6]

Output: 12

```python
# Write your answer here

def sum_of_even(numbers):
    return sum(num for num in numbers if num % 2 == 0)

# Example
numbers = [1, 2, 3, 4, 5, 6]
print(sum_of_even(numbers))  
```

# Create a function that takes a list as input and return reversed list

## Example
lst: [1, 2, 3]

Output: [3, 2, 1]

```python
# Write your answer here

def reverse_list(lst):
    return lst[::-1]

# Example usage
lst = [1, 2, 3]
print(reverse_list(lst))  # Output: [3, 2, 1]
```
