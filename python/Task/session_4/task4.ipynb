```python

def thanos_project(folder_path, num_files=10):
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"Folder created: {folder_path}")
    else:
        print(f"Folder exists: {folder_path}")

    for i in range(num_files):
        file_path = os.path.join(folder_path, f"file_{i}.txt")
        with open(file_path, 'w') as f:
            f.write(f"This is file {i}")
    print(f"{num_files} files created.")

    files = os.listdir(folder_path)
    print(f"Total files before deletion: {len(files)}")

    half_count = len(files) // 2
    files_to_delete = random.sample(files, half_count)
    for file in files_to_delete:
        os.remove(os.path.join(folder_path, file))
    print(f"Deleted {half_count} files randomly.")

    remaining_files = os.listdir(folder_path)
    print(f"Files remaining after deletion: {len(remaining_files)}")
    print("Remaining files:", remaining_files)
```

```python
thanos_project("thanos_folder", num_files=10)
```
